```python
from nltk.tokenize import RegexpTokenizer
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
tokenizer = RegexpTokenizer('\w+')
stopword = set(stopwords.words('english'))
```

```python
ps = PorterStemmer()
```

```python
def clean_review(review):
    
    review = review.lower()
    review = review.replace("<br /><br />"," ")
    
    #Tokenize
    tokens = tokenizer.tokenize(review)
    new_tokens = [token for token in tokens if token not in stopword]
    stemmed_tokens = [ps.stem(token) for token in new_tokens]
    
    cleaned_review = ' '.join(stemmed_tokens)
    
    return cleaned_review
```

```python
x_clean = [clean_review(sent) for sent in x]
x_test_clean = [clean_review(sent) for sent in x_test]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
x_vec = cv.fit_transform(x_clean)
print(cv.get_feature_names())

print(x_vec)
print(x_vec.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB,BernoulliNB, GaussianNB
```

```python
mb = MultinomialNB()
mb.fit(x_vec,y)
```

```python
xt_vec = cv.transform(x_test_clean)
print(xt_vec.shape)
```

```python
cv.get_feature_names()
```

```python
mb_res = mb.predict(xt_vec)
```

```python
bn = BernoulliNB(binarize=0.0)
bn.fit(x_vec,y)
```

```python
bn.predict_proba(xt_vec)
```

```python
bn_res = bn.predict(xt_vec)
```

```python
mb_res = mb_res.tolist()
```

```python
print(len(mb_res))
```

```python
ans_mb = []
for i in range(len(mb_res)):
    if(mb_res[i]==1):
        ans_mb.append('pos')
    else:
        ans_mb.append('neg')
```

```python
ans_bn = []
for i in range(len(bn_res)):
    if(bn_res[i]==1):
        ans_bn.append('pos')
    else:
        ans_bn.append('neg')
```
