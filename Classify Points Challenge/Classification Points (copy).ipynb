```python
dfx = np.array(pd.read_csv("Training Data/Logistic_X_Train.csv"))
dfy = np.array(pd.read_csv("Training Data/Logistic_Y_Train.csv"))
```

```python
print(dfx.shape)
print(dfy.shape)
```

```python
XT = dfx
YT = dfy
```

```python
Xt = np.array(pd.read_csv("Test Cases/Logistic_X_Test.csv"))
```

```python
print(Xt.shape)
```

```python
mean = XT.mean(axis=0)
std = XT.std(axis=0)

# Data Normalisation Step
XT = (XT-mean)/std
Xt = (Xt-mean)/std
```

```python
# Build Our Neural Network Model
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
from keras import backend as K

def r2_score(y_true,y_pred):
    
    
    num = K.sum(K.square(y_true-y_pred))
    denom = K.sum(K.square(y_true- K.mean(y_true)))
    return (1-num/denom)

def getModel():
    model = Sequential()
    model.add(Dense(64,activation='relu',input_shape=(2,)))
    model.add(Dense(32,activation='relu'))
    model.add(Dense(1))
    
    model.compile(optimizer='adam',loss='mse',metrics=[r2_score])
    return model
```

```python
model = getModel()
model.summary()
hist = model.fit(XT,YT,verbose=1,epochs=1000)
```
